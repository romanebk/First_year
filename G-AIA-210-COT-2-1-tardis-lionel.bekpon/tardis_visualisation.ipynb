```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

db = pd.read_csv('cleaned_dataset.csv', sep=",")
db.columns.to_list
```

```python
sns.histplot(db['Pct delay due to rolling stock'], bins=30, kde=True)
sns.histplot(db['Pct delay due to infrastructure'], bins=30, kde=True)
sns.histplot(db['Pct delay due to external causes'], bins=30, kde=True)
sns.histplot(db['Pct delay due to traffic management'], bins=30, kde=True)
sns.histplot(db['Pct delay due to station management and equipment reuse'], bins=30, kde=True)
sns.histplot(db['Pct delay due to passenger handling (crowding, disabled persons, connections)'], bins=30, kde=True)
plt.title("Distribution des retards")
plt.xlabel("Minutes de retard")
plt.ylabel("Fréquence")
plt.show()
```

```python
sns.boxplot(x='Departure station', y='Pct delay due to rolling stock', data=db)
# plt.xticks(rotation=90)
plt.title("Retards par gare de départ")
plt.xlabel("Gare de départ")
plt.ylabel("Minutes de retard")
plt.show()
```

```python
correlation = db.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Corrélations entre variables")
plt.show()
```
